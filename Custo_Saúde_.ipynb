```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
```

```python
df_23 = pd.read_csv('/content/drive/MyDrive/2023.csv', engine='python', sep=';')
df_24 = pd.read_csv('/content/drive/MyDrive/2024.csv', engine='python', sep=';')
df_25 = pd.read_csv('/content/drive/MyDrive/2025.csv', engine='python', sep=';')
```

```python
df = pd.concat([df_23, df_24, df_25], ignore_index=True)
```

```python
df.head()
```

```python
df['modalidade_compra'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['nome_instituicao'] = df['nome_instituicao'].fillna('Não Informado')
df['generico'] = df['generico'].fillna('Não Informado')
df['unidade_medida'] = df['unidade_medida'].fillna('Não Informado')
df['unidade_fornecimento_capacidade'] = df['unidade_fornecimento_capacidade'].fillna('Não Informado')
df['unidade_fornecimento'] = df['unidade_fornecimento'].fillna('Não Informado')
```

```python
df.isnull().sum()
```

```python
df_grup1 = df.groupby(['uf', 'tipo_compra'])[['preco_total','qtd_itens_comprados']].sum().reset_index()
df_grup1 = df_grup1.sort_values(by='preco_total', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=df_grup1, x='uf', y='preco_total', hue='tipo_compra')
plt.title('preco_total por estado')
plt.xlabel('estado (UF)')
plt.ylabel('preco_total')
formatter = mticker.FuncFormatter(lambda x, _: '{:,.0f}M'.format(x/1000000))
plt.gca().yaxis.set_major_formatter(formatter)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_grup1 = df_grup1.sort_values(by='qtd_itens_comprados', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=df_grup1, x='uf', y='qtd_itens_comprados', hue='tipo_compra')
plt.title('qtd de itens comprados por estado')
plt.xlabel('estado (UF)')
plt.ylabel('qtd_itens_comprados')
formatter = mticker.FuncFormatter(lambda x, _: '{:,.0f}M'.format(x/1000000))
plt.gca().yaxis.set_major_formatter(formatter)
plt.tight_layout()
plt.show()
```
